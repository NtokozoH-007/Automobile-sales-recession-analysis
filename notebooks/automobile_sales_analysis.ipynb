## Project Introduction

The automobile industry is highly sensitive to changes in economic conditions. During periods of recession, consumers often reduce discretionary spending, postpone vehicle purchases, and prioritize essential expenses. These behavioral changes directly influence automobile manufacturers, dealerships, and marketing strategies.

This project investigates historical automobile sales data to understand how economic recessions affect vehicle sales performance, advertising expenditure, unemployment, and consumer confidence.

Using Python and modern data visualization techniques, this analysis aims to identify meaningful trends and transform raw data into actionable business insights that can support strategic decision-making.

This portfolio project demonstrates practical skills in exploratory data analysis (EDA), business intelligence, and data storytelling using Python.

## Project Objectives

The primary objectives of this project are to:

- Understand historical automobile sales trends.
- Compare sales performance during recession and non-recession periods.
- Identify which vehicle categories are most affected during recessions.
- Explore the relationship between unemployment and automobile sales.
- Analyze changes in advertising expenditure during economic downturns.
- Visualize regional sales performance using an interactive geographic map.
- Provide business recommendations based on data-driven insights.

## Business Questions

This project seeks to answer the following business questions:

1. How have automobile sales changed over time?
2. How do recession periods compare with non-recession periods?
3. Which vehicle types experience the greatest decline during recessions?
4. Does unemployment significantly influence vehicle sales?
5. How does advertising expenditure change during recession periods?
6. Which regions generate the highest sales during recessions?
7. What strategies can automobile companies implement to reduce the impact of future recessions?
